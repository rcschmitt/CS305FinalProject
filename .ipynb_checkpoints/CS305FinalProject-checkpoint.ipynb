```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv("heart.csv")
data.head()
```

```python
# Get all columns except the label
X = data.iloc[:,0:-1]

# Get the labels
y = data["target"]

#Split into train/test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#Hassanat Distance calculation

def hassanat_dist_metric(df1, df2):
    dist_list = []
    total = 0
    
    for x in range(len(df1)):
        data1 = np.array(df1)[x]
        data2 = np.array(df2)[x]
        
        min_ = min(data1, data2)
        max_ = max(data1, data2)
        
        if min_ >= 0:
            dist = 1-( (1+min_)/(1+max_) )
            dist_list.append(dist)

        else:
            dist = 1-( (1+min_ + np.abs(min_))/(1+max_+np.abs(min_) ) )
            dist_list.append(dist)
    
    total = np.sum(dist_list)
    return total
```

```python
model = KNeighborsClassifier(n_neighbors = 2, metric= hassanat_dist_metric)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
precision = precision_score(y_test, y_pred)
print(precision)
```

```python
diabetes = pd.read_csv("diabetes.csv")
diabetes.head()
```

```python
# Get all columns except the label
X = diabetes.iloc[:,0:-1]

# Get the labels
y = diabetes["Outcome"]

#Split into train/test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = KNeighborsClassifier(n_neighbors = 2, metric= hassanat_dist_metric)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
precision = precision_score(y_test, y_pred)
print(precision)
```
